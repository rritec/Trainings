## 2. Exercise 2: Working with web data without saving locally
* Store into python dataframe
* Download cloud csv file --> df

```python
import matplotlib.pyplot as plt;
import pandas as pd # Import packages
# Assign url of file: url
url = 'https://archive.ics.uci.edu/ml/machine-learning-databases/wine-quality/winequality-red.csv'
df = pd.read_csv(url, sep=';') # Read file into a DataFrame: df
df.head() # Print the head of the DataFrame
```

```python
# Plot first column of df
pd.DataFrame.hist(df.iloc[:, 0:1])
plt.xlabel('fixed acidity (g(tartaric acid)/dm$^3$)')
plt.ylabel('count')
plt.show()
```
