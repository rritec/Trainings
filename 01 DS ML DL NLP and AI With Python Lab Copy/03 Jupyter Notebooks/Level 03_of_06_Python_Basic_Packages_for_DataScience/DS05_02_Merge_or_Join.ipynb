```python
pd.merge(df1,
         df2,
         how = "inner",
         left_on="DEPTNO",
         right_on="DEPTID") # equi join
```

## 2. Exercise 2: Right outer join two DataFrames

```python
pd.merge(left = df1,
         right = df2,
         how = "right", # right outer join # from right side table all we will get irrespective of join pass/fail
         left_on="DEPTNO",
         right_on="DEPTID")
```
