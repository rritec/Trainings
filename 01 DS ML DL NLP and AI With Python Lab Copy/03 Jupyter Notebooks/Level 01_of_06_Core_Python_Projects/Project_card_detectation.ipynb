## Business Case:

Detect the card type (Credit card or Debit card) based on the card number.

###  Note:

*All credit card numbers follow a standard formula by which the first six digits of the credit card identify the company that issued the card. The rest of the digits in the credit card make up the individual account number, with the exception of the last digit, which helps computer programs identify invalid credit card numbers. Determine the type of credit card by matching its first six digits against an industry list.*

Referance Site: https://www.sapling.com/7187438/determine-card-type-card-number

## Solution:
-----------------------------

```python
card = input("Enter your Credit Card number:")

if card[:1] == '4':
    print("Visa")
elif card[:4] == '4026' or card[:6] == '417500' or card[:4] == '4405' or card[:4] == '4508' or card[:4] == '4844' or card[:4] == '4913' or card[:4] == '4917':
    print('Visa Electron')
elif card[:2] == '50' or card[:2] == '51' or card[:2] == '52' or card[:2] == '53' or card[:2] == '54' or card[:2] == '55' :
    print('Mastercard')
elif card[:4] == '5018' or card[:4] == '5020' or card[:4] == '5038' or card[:4] == '5612' or card[:4] == '5893' or card[:4] == '6304' or card[:4] == '6759' or card[:4] == '6761' or card[:4] == '6762' or card[:4] == '6763' or card[:4] == '0604' or card[:4] == '6390':
    print('Maestro')
elif card[:4] == '6304' or card[:4] == '6706' or card[:4] == '6771' or card[:4] == '6709':
    print('Laser')
elif card[:4] == '6011' or card[:3] == '644' or card[:3] == '649' or card[:2] == '65':
    print('Discover Card')
elif card[:2] == '34' or card[:2] == '37':
    print('American Express')
elif card[:3] == '300' or card[:3] == '305' or card[:3] == '309' or card[:2] == '36' or card[:2] == '38' or card[:2] == '39':
    print('Diners Club International')
else:
    print('Others')
```
