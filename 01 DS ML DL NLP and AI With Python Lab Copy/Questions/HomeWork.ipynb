```python
10-Sep-2019
Name: Rishikesh
Problem Statement: Write a function/code to calculate slope, intercept and predicted value of dependent variable Y 
  
# define one  list containing x as independent variable
# define other list containing y as dependent or predicted variable

x=[1000,4000,5000,4500,3000,4000,9000,11000,15000,12000,7000,3000]
y=[9914,40487,54324,50044, 34719, 42551,94871,118914,158484,131348, 78504, 36284]

def UDF_FindSlopeIntercept(l1,l2):
    '''
        Define User defined function to find out slope and intercept
        This function accepts dependent(y) and independent variables (x)
    '''    
    import numpy as np
    
    if len(l1) != len(l2):
        return('Number of elements in both lists are not same !')
    xMean = np.mean(x)
    yMean = np.mean(y)
   
    l3=[]
    l4=[]
    l5=[]

    for i in range(0,len(l1)):
        l3.append((xMean - l1[i]) ** 2)
        l4.append(yMean - l2[i])
        l5.append(  (xMean-l1[i]) *(yMean - l2[i]) ) 
    
    l3Sum = sum(l3)
    l5Sum = sum(l5)
    
    slope = l5Sum/l3Sum
    
    intercept = yMean - (slope * xMean)
    
    return (slope, intercept)


def UDF_predictDV(slope,intercept,l1,l2):
    '''
        Define a user defined a function which will calculate Y or dependent variable and residual error
        This function will accpet slope, intercept, independent variable(x) and dependent variable(y) list
    '''
    l6=[]
    totalErr=0
    errorSS = 0
    for i in range(0,len(l1)):
        #y=mx+c
        l6.append(slope * l1[i] + intercept)
        error = l1[i] - l6[i]
    errorSS = (error ** 2) + totalErr 
    return(l6, errorSS)  


# Calling the function

slope,intercept = UDF_FindSlopeIntercept(x,y)
dv,err  = UDF_predictDV(slope,intercept,x,y)
print('Slope/Gradient       : ', slope)
print('intercept            : ', intercept)
print('Y valuess            : ', dv)
print('Error/Residual Error :',err)

output:
Slope/Gradient       :  10.622195457815229
intercept            :  1383.4713801253674
Y valuess            :  [12005.666837940596, 43872.25321138628, 54494.44866920151, 49183.350940293894, 33250.057753571054, 43872.25321138628, 96983.23050046242, 118227.62141609288, 160716.4032473538, 128849.81687390812, 75738.83958483196, 33250.057753571054]
Error/Residual Error : 915065994.0943843

        
```

```python
#Name: Rishikesh
#Date: 11-Sep-2019
#Problem Statement: Write a function to find R2SCore
    
def FindR2Score(ytlist,yplist):
    '''
    Returns R2Score 
    Aguments: 2 arguments
        1. YTruh Real values
        2. yPredicted values by machine
    '''
    import numpy as np
    
    RSS = np.sum(np.square(np.array(ytlist) - np.array(yplist)))
    SSTot = np.sum(np.square(np.array(yplist) - np.mean(ytlist)))
    
    return (1 - (RSS/SSTot))
    
# Drive script
x=[3,-0.5,2,7]
y=[2.5,0,2,8]
print(FindR2Score(x,y))
```

```python
Name:prasanna karra 09/24/2019

 use the dot function to compute inner products of vectors, to multiply a vector by a matrix, and to multiply matrices.
  
  dot is available both as a function in the numpy module and as an instance method of array objects:

import numpy as np

x = np.array([[1,2],[3,4]])
y = np.array([[5,6],[7,8]])

v = np.array([9,10])
w = np.array([11, 12])

# Inner product of vectors; both produce 219
print(v.dot(w))
print(np.dot(v, w))

# Matrix / vector product; both produce the rank 1 array [29 67]
print(x.dot(v))
print(np.dot(x, v))

# Matrix / matrix product; both produce the rank 2 array
# [[19 22]
#  [43 50]]
print(x.dot(y))
print(np.dot(x, y))
```
