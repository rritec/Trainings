## Visualize Data

```python
# Histogram
sns.distplot(tips.total_bill, hist=True, kde=False);
```

```python
# KDE
sns.distplot(tips.total_bill, hist=False, kde=True);
```

```python
# Histo + KDE
sns.distplot(tips.total_bill, hist=True, kde=True);
```

#### Pair Plot w/Scatter & Histogram

```python
sns.pairplot(tips);
```

```python
sns.pairplot(tips, hue='smoker');
```

```python
count = 0
for i in range(len(tips)):
    if tips['smoker'][i] == 'Yes':
        count += 1
print('Smoker is {0:.3}% of total.'.format(count*100/len(tips)))
```

#### Bivariate Scatter Plot

```python
sns.jointplot(x="total_bill", y="tip", data=tips);
```

#### Hex Plot

```python
sns.jointplot(x='total_bill', y='tip', kind="hex", color="k", data=tips);
```

```python
sns.jointplot(x="total_bill", y="tip", data=tips, kind="kde");
```

#### Categorical Scatterplot

```python
sns.stripplot(x="size", y="total_bill", data=tips);
```

#### Categorical w/Jitter

```python
sns.stripplot(x="size", y="total_bill", data=tips, jitter=True);
```

#### Swarm Plot
Positions each scatterplot point on the categorical axis with an algorithm that avoids overlapping points

```python
sns.swarmplot(x="size", y="total_bill", data=tips);
```

```python
sns.swarmplot(x="size", y="total_bill", hue="sex", data=tips);
```

#### Boxplot

```python
sns.boxplot(x="day", y="total_bill", hue="time", data=tips);
```

```python
tips["weekend"] = tips["day"].isin(["Sat", "Sun"])
sns.boxplot(x="day", y="total_bill", hue="weekend", data=tips, dodge=False);
```

```python
sns.factorplot(x="time", y="total_bill", hue="smoker",
               col="day", data=tips, kind="box", size=4, aspect=.5);
```

#### Violin Plot

```python
sns.violinplot(x="day", y="total_bill", hue="sex", data=tips, split=True);
```
