```python
%matplotlib inline

import matplotlib.pyplot as plt
import numpy as np
from numpy.fft import fft2
from scipy.misc import imread
import os
```

## Get Pictures Function 

```python
def get_pics(path):
    files = os.listdir(path)
    for i, file in enumerate(files):
        if file == '.DS_Store':
            files.pop(i)
    return files
```

```python
def process_pics(path, names=[], flatten=True):
    pics = []
    for name in names:
        pics.append( imread(path + name, flatten=flatten) )
    return pics
```

```python
def fft_process(img):
    f = np.fft.fft2(img)
    fshift = np.fft.fftshift(f)
    return 20*np.log(np.abs(fshift))
```

## Load Pics

```python
path = '/Users/davidziganto/images/'
files = get_pics(path)
pics = process_pics(path, names=files, flatten=True)
```

```python
for pic in pics[:10]:
    plt.subplot(121)
    plt.imshow(pic, cmap='gray')
    plt.subplot(122)
    plt.imshow(fft_process(pic), cmap='gray')
    plt.show()
```
