```python
%matplotlib inline
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
def bagging_simulator(n_rows=10**4, n_runs=10**4):
    '''returns list of percentages indicating how many indices selected for bagging'''
    idx = np.arange(n_rows)
    choice_list = []
    
    for i in range(n_runs):
        choices = np.random.choice(idx, size=n_rows, replace=True)
        pct = len(set(choices)) / len(idx)
        choice_list.append(pct)    
    
    return choice_list
```

```python
in_bag = bagging_simulator()
sns.distplot(in_bag, bins=15)
plt.title('% In-Bag');
```
