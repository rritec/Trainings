```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
```

```python
def sigmoid(x):
    return 1 / (1 + np.e ** -x)
```

```python
x = np.linspace(-15, 15, 100)
lst = [sigmoid(i) for i in x]
```

```python
plt.plot(x, lst);
plt.axhline(0.5, c='k', alpha=0.5)
plt.vlines(0, 0, 1, alpha=0.5)
plt.title('Sigmoid Function')
plt.xlabel('x')
plt.ylabel('probability');
```
